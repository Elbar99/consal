```python
from tkinter import *
from random import *

p = 6
n = 4
#

grill =  [[0 for j in range(p)] for i in range(n)]


def peut_tomber(i,j):
    if i == n-1: # tout en bas
        return False
    if grill[i+1][j]: # case juste en-dessous
        return False
    if j>0 and grill[i][j-1]: # à gauche
        return False
    if j<p-1 and grill[i][j+1]: # à droite
        return False
    return True


def faire_tomber_un_bloc(j):
    i = 0
    while peut_tomber(i,j):
        i = i + 1
        grill[i][j] = 1
    return i,j


def faire_tomber_des_blocs(k):
    for l in range(k):
        j = randint(0,p-1)
        faire_tomber_un_bloc(j)
        l = l+1
    return

def afficher_grill():
    for i in range(n):
        for j in range(p):
            print(grill[i][j], end="")
        print()
    return
afficher_grill()


########

n = 50 # nb de lignes
p = 100 # nb de colonnes
grill = [[0 for j in range(p)] for i in range(n)]


echelle = 5 # échelle
nb_blocs = 500
root = Tk()
canvas = Canvas(root, width=500, height=300, background="white")
canvas.pack(fill="both", expand=True)
def afficher_tableau():
    canvas.delete("all") 
    for i in range(n):
        for j in range(p):
            if grill[i][j]:
                 canvas.create_rectangle(j*echelle,i*echelle,j*echelle+echelle-1,i*echelle+ echelle-1,width=1,fill='green')
                
    return
 
      
def action_bloc():
    faire_tomber_des_blocs(nb_blocs)
    afficher_tableau()
#     for i in range(nb_blocs):   
#         faire_tomber_un_bloc(i)
        
    return

bouton_bloc = Button(root,text="Afficher blocs", width=20, command=action_bloc)
bouton_bloc.pack(pady=10)
bouton_quitter = Button(root,text="Quitter", width=20, command=root.quit)
bouton_quitter.pack(side=BOTTOM, pady=10)
root.mainloop()



#Arbres browniens)
def est_dedans(i,j):
    pass

def faire_libre(i,j):
    pass

def lancer_un_bloc():
    pass

def lancer_des_blocs(k):
    pass

```
